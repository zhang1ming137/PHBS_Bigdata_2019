## 1.3. Which variables are significant in the model?


### Use T-Test to find whether the features are significant in the model

```python
def T_test(X,Y,cita):
    error = np.sum(np.square(X@cita-Y))
    Cii = np.linalg.inv(X.T@X).diagonal()
    sigma = error /(X.shape[0]-X.shape[1])
    S = np.sqrt(sigma*Cii)
    results = pd.DataFrame()
    results['index'] = df.columns.values[:11]
    results.set_index(['index'], inplace=True)
    results['coefficient'] = np.array(cita)
    t = cita.T / S
    results['t-value']=np.array(t)
    p = stats.t.sf(np.abs(t), (X.shape[0]-X.shape[1]))*2
    results['p-value']=np.array(p)
    return results
M = np.array(Trainset.iloc[:,:11])
N = np.array(Trainset.iloc[:,11])
results = T_test(M,N,cita)
results
```

### The p-value of  Year and CH4 is larger than 0.05 which means they are not significant. That is to say, Month,MEI,CO2,N2O, CFC-11, CFC-12, TSI and Aerosols are siginificant.

# Problem2: Regularization

## 2.1 loss function for linear model with L1 regularization, L2 regularization

### For L1 regularization:

### $$ J(\theta) = \cfrac{1}{2m} [\displaystyle \sum^{m}_{i=1}{(h_\theta(x ^ {(i)})  - y ^ {(i)})^2 }+ \lambda \displaystyle \sum^{n}_{j=1}{| \theta _j |}]  $$


### For L2 regularization:

### $$ J(\theta) = \cfrac{1}{2m} [\displaystyle \sum^{m}_{i=1}{(h_\theta(x ^ {(i)})  - y ^ {(i)})^2 }+ \lambda \displaystyle \sum^{n}_{j=1}{\theta _j ^2}]  $$

## 2.2 L2 regularization

```python
def closed_form_2 (X,Y,lamda):
    cita_2 = np.linalg.inv(X.T@X+lamda*np.identity(X.shape[1]))@X.T@Y
    print("Temp =",cita_2[0],"* Year +",cita_2[1],"Month +",cita_2[2],"* MEI +",cita_2[3],"* CO2 +",cita_2[4],"* CH4 +",cita_2[5],"* N2O +",cita_2[6],"* CFC-11 +",cita_2[7],"* CFC-12 +",cita[8],"* TSI +",cita[9],"* Aerosols +",cita[10])
    return cita_2
```

# Problem 4——Gradient descent algorithm

# The iteration expression:
$ Repeat:  \ \{ $
$$ \theta _j := \theta _j - \alpha \frac{1}{m} \sum^{m}_{i=1}{(h_\theta(x ^ {(i)})-y^{(i)})x^{(i)}_j}$$
$ (simultaneously update \theta _j \  for \ j = 0, ..., n)\}$
$\  for \ j = 0, ..., n $


```python
def gradientDescent(df_x, df_y, alpha, e=1e-17, max_loop=100000):
    iteration = 0
    n = df_x.shape[1]
    m = df_x.shape[0]
    theta = np.random.randn(n)
    flag = 1
    while iteration < max_loop and flag == 1: 
        iteration += 1
        print(iteration)
        last_theta = np.array(theta)
        for j in range(n):
            theta[j] = theta[j] - alpha*sum(((np.array(df_x)@theta)-df_y)*df_x[:, j]) / m
        if sum((theta-last_theta) **2) < e:
            flag = 0
            print(sum((theta-last_theta) **2))
    return theta
```
